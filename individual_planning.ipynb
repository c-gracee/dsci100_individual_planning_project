## Exploratory Data Analysis and Visualization

### Wrangling the data

### Mean value for each quantitative variable in the players.csv file

### Exploratory visualization

- **mean played hours vs subscribe**: From the visualization, I can see that for players who subscribed, their average played hours is higher than those who did not subscribe. This can contribute to answering the broad question of interest where the player behaviours might be predictive of subscription status. In addition, there can be futher investigation on whether played hours is related to experience, consequently contributing to answering the specific question of interest.

*refer to code cell below*

```r
ph_subscribe_plot <- players |>
    group_by(subscribe) |>
    summarise(mean_hours = mean(played_hours, na.rm = TRUE)) |>
    ggplot(aes(x = subscribe, y = mean_hours)) +
    geom_col() +
    labs(
        x = "Subscription status (TRUE/FALSE)",
        y = "Mean played hours (hrs)",
        title = "Average played hours by subscription status")
ph_subscribe_plot
```

- **mean played hours vs experience**: The visualization below shows that the highest mean played hours occurs for regular experienced players. This gives me insights for answering whether mean played hours is different amongst different experience types and hence answer the question of whether experience can predict subscription status.

*refer to code cell below*

```r
ph_exp_plot <- players |>
    group_by(experience) |>
    summarise(mean_hours = mean(played_hours, na.rm = TRUE)) |>
    ggplot(aes(x = experience, y = mean_hours)) +
    geom_col() +
    labs(
        x = "Experience type",
        y = "Mean played hours (hrs)",
        title = "Average played hours by experience type")
ph_exp_plot
```

- **experience vs gender**: The histogram visualization shows that males are the majority players for each experience type, hence potentially influencing the charactersitics of experience type and subscription status.

*refer to code cell below*

```r
gender_exp <- players |>
  select(gender, experience)

gender_exp_plot <- ggplot(gender_exp, aes(x = experience, fill = gender)) +
    geom_bar(position = "fill") +
    scale_y_continuous(labels = scales::percent) +
    labs(
        x = "Experience",
        y = "Proportion of players",
        fill = "Gender",
        title = "Proportion of genders across experience levels")
gender_exp_plot
```

- **experience vs subscribe**: From this visualization, we can see amateur experienced players are the majority who subscribed. Following amateur are veteran, regular, beginner, and pro, respectively. Overall, the visualization allows me to understand the potential experience types that are more likely to subscribe.

*refer to code cell below*

```r
exp_sub <- players |>
    select(experience, subscribe)

exp_sub_plot <- ggplot(exp_sub, aes(x = subscribe, fill = experience)) +
    geom_bar(position = "fill") +
    scale_y_continuous(labels = scales::percent) +
    labs(
        x = "Subscribed (TRUE / FALSE)",
        y = "Proportion of experience",
        fill = "Experience",
        title = "Proportion of experience types by subscription status")
exp_sub_plot
```

## Methods and plan

The method for answering the specific question can be using a logistic regression model. This method is appropriate because subscription status is a logical variable type.

K nearest neighbors or linear regression cannot be used because I cannot classification or regression modes on response variables that are logical. 

For this method, I am assuming that experience influences subscription status. This assumption brings out a weakness to this method such that if there are no correlation or causation, the entire model and prediction can be wrong. 

I do not need to split the data for using logistic regression because this method uses the probability of whether the players of different experiences are subscribed or not, hence requiring the input of experience data into the regression at all times. Consequently, there will be no cross validations preformed.
